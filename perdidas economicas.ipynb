```python
import pandas as pd
df=pd.read_csv('requestedDataFormatted(1).csv',sep=';',encoding='latin-1')
df['anio']=df['Start date'].apply(lambda x: int(x.split('/')[2]))
train=df[(df['anio']==2014) | (df['anio']==2015)|(df['anio']==2016)|(df['anio']==2013)|(df['anio']==2012)|(df['anio']==2011)|(df['anio']==2010)]
xtrain=train.iloc[:,[7,9,10]].values
ytrain=train.iloc[:,[8]].values

xtest1=df[df['anio']==2017].iloc[:,[7,9,10]].values
ytest1=df[df['anio']==2017].iloc[:,[8]].values

xtest2=df[df['anio']==2018].iloc[:,[7,9,10]].values
ytest2=df[df['anio']==2018].iloc[:,[8]].values

xtest3=df[df['anio']==2019].iloc[:,[7,9,10]].values
ytest3=df[df['anio']==2019].iloc[:,[8]].values
```

```python
import math 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
# difference of lasso and ridge regression is that some of the coefficients can be zero i.e. some of the features are 
# completely neglected
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain =scaler.fit_transform(xtrain)
scaler=StandardScaler()
xtest1=scaler.fit_transform(xtest1)
scaler=StandardScaler()
xtest2=scaler.fit_transform(xtest2)
scaler=StandardScaler()
xtest3=scaler.fit_transform(xtest3)
```

```python
lasso = Lasso()
lasso.fit(xtrain,ytrain)
train_score=lasso.score(xtrain,ytrain)
test_score=lasso.score(xtest2,ytest2)
coeff_used = np.sum(lasso.coef_!=0)
```

```python
print ("training score:", train_score) 
print ("test score: ", test_score)
print ("number of features used: ", coeff_used)
```

```python
lasso001 = Lasso(alpha=0.0000001, max_iter=10e5)
lasso001.fit(xtrain,ytrain)
train_score001=lasso001.score(xtrain,ytrain)
test_score001=lasso001.score(xtest1,ytest1)
coeff_used001 = np.sum(lasso001.coef_!=0)
print ("training score for alpha=0.01:", train_score001) 
print ("test score for alpha =0.01: ", test_score001)
print ("number of features used: for alpha =0.01:", coeff_used001)
```
